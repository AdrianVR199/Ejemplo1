```
    import numpy as np
    import matplotlib.pyplot as plt 
  
    # INGRESO
    a = -5 # rango de tiempo
    b = 5
    dt = 0.1
    sigma = 0 # s = 0 + 0j
    omega = 0
    
    # PROCEDIMIENTO
    t =np.arange(a, b, dt)
    s = complex(sigma, omega)
    senal = np.exp(s*t)
    
    #SALIDA 
    plt.figure(1)
    plt.plot(t,np.real(senal), label='real')
    plt.plot(t,np.imag(senal), label='imaginaria')
    plt.xlabel('t')
    plt.ylabel('x(t)')
    plt.legend()
    plt.margins(0.1)
    plt.title('e^('+str(sigma)+'+'+str(omega)+'j)')
    plt.show()
    
    
```

```
import scipy.integrate as integrate

#Parametros
w=1
a=0.1
b=0
c=4*2*np.pi #un periodo
dte= 0.005

# Procediminto
te=np.arange(b,c,dte)
xe=np.exp(-a*te)  #ecuacion a la cual le queremos conocer la energia

#energia en un periodo
cuadradoxe= xe**2 # se elevala funcion al cuadrado
energiaxe = integrate.simps(cuadradoxe,te) # sacamos su integral

#salida
print('la energia de xe es:',energiaxe)

# grafica
plt.plot(te,xe,label='xe')
plt.fill_between(te,0,cuadradoxe,color='pink')
plt.plot(te,cuadradoxe,'g--', label='xe^2')
plt.xlabel('t')
plt.legend()
plt.show()
```
